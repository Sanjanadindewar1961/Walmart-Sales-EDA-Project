```python
import pandas as pd
import numpy as np

df = pd.read_csv("Walmart_clean_data.csv")
```

```python
unit_price_arr = df['unit_price'].to_numpy()
quantity_arr = df['quantity'].to_numpy()

print("Quantity array sample:", quantity_arr[:10])
```

# Square each quantity using NumPy

```python
quantity_squared = quantity_arr ** 2
print(quantity_squared[:10])
```

# Find max, min, sum using NumPy

```python
print("Max quantity:", np.max(quantity_arr))
print("Min quantity:", np.min(quantity_arr))
print("Total quantity sold:", np.sum(quantity_arr))
```

# Filter values using NumPy condition
find all unit prices > 50

```python
high_prices = unit_price_arr[unit_price_arr > 50]
print(high_prices[:10])
```

# Compute correlation between two columns using NumPy
unit price vs quantity

```python
corr = np.corrcoef(unit_price_arr, quantity_arr)
print("Correlation matrix:\n", corr)
```

# Convert quantity to a NumPy array

```python
quantity_arr = df['quantity'].to_numpy()
print(quantity_arr[:10])
```

```python
import numpy as np

for col in ['unit_price', 'quantity', 'rating', 'Total']:
    data = df[col]
    print(f"\n--- {col.upper()} Statistics ---")
    print(f"Mean: {np.mean(data):.2f}")
    print(f"Median: {np.median(data):.2f}")
    print(f"Std Dev: {np.std(data):.2f}")
    print(f"25th Percentile: {np.percentile(data, 25):.2f}")
    print(f"75th Percentile: {np.percentile(data, 75):.2f}")
```
