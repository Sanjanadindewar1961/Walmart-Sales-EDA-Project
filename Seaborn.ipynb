```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Walmart_clean_data.csv")  
```

# Boxplot – Total Sales by Payment Method

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='payment_method', y='Total')
plt.title("Total Sales by Payment Method")
plt.show()
```

# Barplot – Total Sales by Category

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='category', y='Total', estimator=sum)
plt.xticks(rotation=45)
plt.title("Total Sales by Category")
plt.show()
```

# Scatterplot – Unit Price vs Total Sales

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='unit_price', y='Total')
plt.title("Unit Price vs Total Sales")
plt.show()
```

# Countplot – Number of Transactions per Payment Method

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='payment_method')
plt.title("Transaction Count by Payment Method")
plt.show()
```

```python
sns.countplot(data=df, y='category', palette='viridis')
plt.title("Product Category Frequency")
plt.show()
```

```python
sns.countplot(data=df, x='payment_method', palette='magma')
plt.title("Payment Method Usage")
plt.show()
```
